# Binning

```python
df_train['Age_Cat'] = df_train['Age'].apply(lambda x:'child' if x<18 else 'young' if x<50 else 'old')
df_test['Age_Cat'] = df_test['Age'].apply(lambda x:'child' if x<18 else 'young' if x<50 else 'old')

```

```python
df_train.dtypes
```

# Feature Encoding

```python
df_train[['male','female']] = pd.get_dummies(df_train['Sex'])
df_test[['male','female']] = pd.get_dummies(df_test['Sex'])


df_train[['C','Q','S']] = pd.get_dummies(df_train['Embarked'])
df_test[['C','Q','S']] = pd.get_dummies(df_test['Embarked'])
```

```python
df_train[['male','female']] = pd.get_dummies(df_train['Sex'])
df_test[['male','female']] = pd.get_dummies(df_test['Sex'])
```

```python
df_train[['C','Q','S']] = pd.get_dummies(df_train['Embarked'])
df_test[['C','Q','S']] = pd.get_dummies(df_test['Embarked'])
```

```python
df_train['Age_Cat'] = df_train['Age_Cat'].map({'child':0,'young':1,'old':2})
df_test['Age_Cat'] = df_test['Age_Cat'].map({'child':0,'young':1,'old':2})
```
